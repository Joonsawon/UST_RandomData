```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,10,0.1)
f = 0.1
def sin(k,t):
    return np.sin(2*np.pi*k*f*t)
    
a = sin(1,t)
b = sin(2,t)
c = sin(3,t)
x = a + b + c

plt.figure(figsize=(8,2))
plt.plot(t,x)
plt.grid(True)

plt.figure(figsize=(8,2))
#plt.subplot(311)
plt.plot(t,a)
plt.grid(True)
plt.figure(figsize=(8,2))
#plt.subplot(312)
plt.plot(t,b)
plt.grid(True)
plt.figure(figsize=(8,2))
#plt.subplot(313)
plt.plot(t,c)
plt.grid(True)
```
